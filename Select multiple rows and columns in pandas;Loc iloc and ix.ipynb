```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo
```

```python
ufo.head()
```

```python
#row zero the column means all column
ufo.loc[0,:]
```

```python
#selecting multiple row column
ufo.loc[[0, 1, 2],:]
```

```python
#another way of selecting  multiple row
#selecting multiple row 
ufo.loc[0:2,:]
```

```python
#selecting a column
ufo.loc[:, 'City']
```

```python
#selecting multipiple columns
ufo.loc[:, ['City', 'State']]
```

```python
#another way of selecting  multiple column
#selecting multiple column
ufo.loc[:,'City': 'State']
```

```python
#row 0 to two columns City to state
ufo.loc[0:2,'City': 'State']
```

# using loc with boolean 

```python
ufo.loc[ufo.City=='Eagle River', 'State']
```

```python
#in iloc in this example it is inclusive of the 0 and exclusive of the last number
ufo.iloc[:, 0:4]
```

```python
ufo.iloc[0:3, :]
```
